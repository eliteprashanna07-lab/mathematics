```python
def dot_product(vec1, vec2):
  vec1= vec1.flatten()
  vec2= vec2.flatten()
  return np.sum(np.multiply(vec1, vec2))

def matrix_vector_product(mat, vec):
  m,n= mat.shape
  p,q=vec.shape

  assert n==p, "the matrix and vectors are not compatible for multiplication!"

  result = np.zeros((m,q), dtype=float)
  for i in range(m):
    result[i] = dot_product(mat[i,:], vec)
    return result
if __name__ =="__main__":

  mat1 = np.full((3,3),3,dtype=int)
  vec1 =np.full((3,1),2,dtype=int)
  print("Matrix= ")
  print(mat1)
  print("Vector= ")
  print(vec1)
  print("Result=")
  result1= matrix_vector_product(mat1, vec1)
  print(result1)

  result2= mat1@vec1
  print("Result=")
  print(result2)
```
